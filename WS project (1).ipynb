```python
record=pd.DataFrame({})
record["PLAYER NAME"]=player
record["SPAN"]=span
record["MATCHES"]=matches
record["INNINGS"]=inns
record["NOT OUTS"]=no
record["RUNS"]=runs
record["HIGHEST SCORE"]=hs
record["AVERAGE"]=avg
record["BALL FACED"]=bf
record["STRIKE RATE"]=sr
record["HUNDREDS"]=hun
record["FIFTIES"]=fif
record["DUCKS"]=ducks
record["FOURS"]=fours
record["SIXES"]=sixes
record
```

```python
import pymysql as mycon
```

```python
con1=mycon.connect(host="localhost",user="root",database="record")

print("connection established successfully")
```

```python
mycursor=con1.cursor()



mycursor.execute("create table if not exists cric(PLAYER varchar(50),SPAN int,MATCHES int,INNINGS int,NO int,RUNS int,HS int,AVERAGE float,BF int,SR float,HUNDREDS int,FIFTIES int,DUCKS int,FOURS int,SIXES int)")
print("successfully created table")
```

```python
for i in range(len(player)):
    mycursor.execute("""insert into cric values("%s","%s","%s","%s","%s","%s","%s","%s","%s","%s","%s","%s","%s","%s","%s")"""%(player[i],span[i],matches[i],inns[i],no[i],runs[i],hs[i],avg[i],bf[i],sr[i],hun[i],fif[i],ducks[i],fours[i],sixes[i]))
    con1.commit()
```
